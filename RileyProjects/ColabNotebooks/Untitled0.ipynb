```python
import requests
import datetime
from google.colab import drive, userdata

# 1. Connect to your Drive
drive.mount('/content/drive')

# 2. Setup your keys
API_KEY = userdata.get('ELEVEN_LABS_KEY') # This pulls the key you just added
VOICE_ID = "pNInz6obpgnuMvscL7qy" # High-quality Masculine American voice (Josh)
FILENAME = f"/content/drive/My Drive/AI_Brief_{datetime.date.today()}.mp3"

# 3. The Daily Brief Content (I will provide this daily)
transcript = """
Good morning, Graham. It’s February 1st, 2026.
Yesterday, the industry shifted from 'chat' to 'agency'.
Elon Musk is pushing algorithm transparency on X,
while Nate B. Jones warns that 2026 is the year of context memory.
You're driving north—keep an eye on the transition from models to systems.
"""

# 4. Generate & Save
url = f"https://api.elevenlabs.io/v1/text-to-speech/{VOICE_ID}"
data = {"text": transcript, "model_id": "eleven_multilingual_v2"}
response = requests.post(url, json=data, headers={"xi-api-key": API_KEY})

with open(FILENAME, 'wb') as f:
    f.write(response.content)

print(f"✅ Success! Your brief is ready: {FILENAME}")
```
